```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
temp_vendas = {'Temperatura': [37, 40, 49, 61, 72, 79, 83, 81, 75, 64, 53, 42],
               'Vendas': [292, 228, 324, 376, 440, 496, 536, 556, 496, 412, 324, 320]}

df = pd.DataFrame(temp_vendas)
# convertendo para graus celsius
df['Temperatura'] = (df['Temperatura'] - 32) * (5 / 9)
df = df.sort_values('Temperatura')
df
```

```python
plt.title('Ice Cream Truck')
plt.plot(df['Temperatura'], df['Vendas'], 'r--', label='US$', marker='.')

plt.ylabel('Vendas (US$)')
plt.xlabel('Temperatura (Celsius')

plt.legend()
plt.savefig('icecream_sales.pdf')
plt.show()
```
