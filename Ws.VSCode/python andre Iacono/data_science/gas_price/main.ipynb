```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gasolina = pd.read_csv('gas_prices.csv')
gasolina
```

```python
'''
criando grafico
'''
plt.title('Valor da gasolina em US$')

#adicionando paises manualmente
# plt.plot(gasolina['Year'], gasolina['USA'],'g.-', label='USA')
# plt.plot(gasolina['Year'], gasolina['France'], 'b.-', label='France')
# plt.plot(gasolina['Year'], gasolina['Canada'], 'r.-', label='Canada')

# adicionadno os paises com For
lista_paises = ['USA', 'Canada', 'France', 'Italy']
for pais in gasolina:
    if pais in lista_paises:
        plt.plot(gasolina['Year'], gasolina[pais], label=[pais], marker='.')

plt.ylabel('Preos da gasolina')
plt.xlabel('Anos')

# plt.axis((1990, 2008, 0, 8))
# plt.xlim(1990, 2008)
plt.ylim(1, 8)
plt.xticks(gasolina['Year'][::2])

plt.legend()
plt.savefig('Gasolina_paises.pdf')
plt.show()
```
