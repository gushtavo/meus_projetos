ANALISANDO DADOS

```python
import pandas as pd
import re

casas = pd.read_csv('house.csv')

casas.info()
```

```python
'''
filtrando dados com condições 
'''

filtro_casa = casas.loc[(casas['Rooms'] == 3) & (casas['Type'] == 'h') & (casas['Price'] > 500000)]
filtro_casa
```

```python
'''
filtrando dados com Regular Expresssion

#str.contains => função para definir uma string/texto como parametro
# re.I => flag para executar independente da forma literal digitada como parametro
'''

filtro_rua = casas.loc[casas['Address'].str.contains('turner sT|Turner rd', flags=re.I)] 
filtro_rua
```

```python
'''
filtrando dados com Regular Expresssion e condições

regx e conditions
'''

filtro_rua_casa = casas.loc[casas['Address'].str.contains('^59', flags=re.I) & (casas['Price'] <= 500000)]
filtro_rua_casa
```
