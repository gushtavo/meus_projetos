```python
import pandas as pd 

microdadosFifa = pd.read_csv('fifa.csv')

microdadosFifa = microdadosFifa.drop(columns=['Photo', 'Flag','Club Logo'])

microdadosFifa.head()
```

```python
'''
Funções Loc e Sorted
'''

microdadosFifa.loc[microdadosFifa['Nationality'] == 'Brazil'].sort_values(by='Age')
```

```python
# Adionando coluna

microdadosFifa['Total'] = microdadosFifa['Acceleration'] + microdadosFifa['Agility'] + microdadosFifa['Reactions']

microdadosFifa = microdadosFifa[['Name', 'Total']]
microdadosFifa = microdadosFifa.sort_values('Total', ascending=False)
microdadosFifa
```

```python
# Salva os arquivos
microdadosFifa.to_csv('microdadosFifa.csv', index=False)
```
