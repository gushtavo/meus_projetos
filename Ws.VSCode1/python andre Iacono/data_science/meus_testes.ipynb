```python
import pandas as pd

microdadosfifa = pd.read_csv('fifa.csv/fifa.csv')

microdadosfifa.info()
```

```python
clube_nome = microdadosfifa[['Name', 'Club']]

display(clube_nome)
```

```python
acima_25 = microdadosfifa.loc[microdadosfifa['Age'] > 25, ['Name', 'Age']].sort_values(by='Age')

acima_25.head(10)
```

```python
overall = microdadosfifa[['Overall', 'Name']]

overall.head(10)
```

```python
penaltis = microdadosfifa[['Penalties', 'Name']]

penaltis
```

```python
colunasSelecionadas = ['Name', 'Club', 'Age', 'Nationality']

nacionalidades = microdadosfifa.filter(items=colunasSelecionadas)

brasileros = nacionalidades.loc[(nacionalidades['Nationality'] == 'Brazil') & (nacionalidades['Age'] > 20)]

brasileros.head(15)
```
